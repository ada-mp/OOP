## 1. Execute as células abaixo e explique os resultados

```python
class A:
    def f(self):
        print('f of A')
        
    def g(self):
        print('g of A')
        self.f()

a = A()
a.f()
a.g()
```

```python
class B:
    def __init__(self, x):
        print(f'New B with {x}')
        self.x = x
        
    def f(self, y):
        print(f'f for a B with {self.x}')
        return self.x + y
    
b1 = B(4)
b2 = B(7)
print(f'{b1.f(10)}')
print(f'{b2.f(20)}')
```

```python
class C:
    def f(self):
        print('f of C')
    def g(self):
        print('g of C')
        self.f()

class D(C):
    def f(self):
        print('f of D')
              
c = C()
d = D()
c.f()
c.g()
d.f()
d.g()
```

```python
class E:
    def __init__(self, x):
        self.x = x
    def f(self):
        return self.x

class F(E):
    def f(self):
        return 2 * self.x
    
e = E(12)
f = F(22)
print(e.f())
print(f.f())
```

```python
class G:
    def __init__(self, x):
        self.x = x
    def f(self):
        return self.x
    
class H(G):
    def __init__(self, x):
        self.x = 3 * x
        
g = G(15)
h = H(35)
print(g.f())
print(h.f())
```

```python
class I:
    def __init__(self, x):
        self.x = x
    def f(self):
        return self.x
    def g(self):
        return 10 * self.f()
    
class J(I):
    def __init__(self, x):
        self.x = x
    def g(self):
        return 30 * self.f()
    
i = I(3)
j = J(11)
print(i.f())
print(i.g())
print(j.g())
```

```python
class AA:
    def __init__(self, x):
        self.x = x
    def f(self, y):
        return self.x + y
    def g(self, y):
        return 2 * self.f(y)
    
class BB(AA):
    def f(self, y):
        return self.x - y

class CC(AA):
    def g(self, y):
        return 3 * self.f(y)
    
class DD(BB):
    def g(self, y):
        return 5 * self.f(y)
    
class EE(CC):
    def f(self, y):
        return self.x + 2 * y
    
class FF(CC):
    def __init__(self, x):
        CC.__init__(self, 5 * x)
        
aa = AA(7)
bb = BB(11)
cc = CC(13)
dd = DD(17)
ee = EE(19)
ff = FF(23)
print(aa.g(2))
print(bb.g(2))
print(cc.g(2))
print(dd.g(2))
print(ee.g(2))
print(ff.g(2))
```

## 2. Implemente a classe pedida

Temos a seguinte classe disponível que implementa um ponto num espaço tridimensional:

```python
class Point:
    def __init__(self, x, y, z):
        self._coords = x, y, z
    def x(self):
        return self._coords[0]
    def y(self):
        return self._coords[1]
    def z(self):
        return self._coords[2]

class Particle(Point):
  def __init__(self,x,y,z,mass):
    self._coords = x,y,z
    self._mass = mass
  def mass(self):
    return self._mass
```

Usando herança, crie uma classe denominada `Particle`, que tem além da posição da partícula a sua massa; a massa é fornecida junto com as coordenadas na criação do objeto. Essa classe deve ter um método denominado `mass` que retorna o valor da massa da partícula.

Veja o exemplo de uso abaixo:

```python
p1 = Particle(1.0, 2.0, -1.0, 100.) # partícula em (1, 2, -1) com massa 100
print(f'A particle with mass {p1.mass()} was detected at ({p1.x()}, {p1.y()}, {p1.z()})')
```
